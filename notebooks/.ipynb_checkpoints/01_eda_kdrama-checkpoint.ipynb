```python
import pandas as pd
import re
from pathlib import Path

# Base paths
# Notebook is inside Kdrama_analytics/notebooks → so go one level up
BASE_DIR = Path.cwd().parent
DATA_DIR = BASE_DIR / "data"

KAGGLE_PATH = DATA_DIR / "kdrama_kaggle_1500.csv"
MY_RATINGS_PATH = DATA_DIR / "my_kdrama_ratings.csv"

KAGGLE_PATH, MY_RATINGS_PATH
```

```python
# Load the Kaggle K-drama dataset
kdrama = pd.read_csv(KAGGLE_PATH)

print("Kaggle dataset shape:", kdrama.shape)
kdrama.head(3)
```

```python
# Load your personal ratings
my_ratings = pd.read_csv(MY_RATINGS_PATH)

print("My ratings shape:", my_ratings.shape)
my_ratings.head()
```

```python
# Clean and rename Kaggle dataset columns
kdrama = kdrama.rename(columns={
    "Name": "title",
    "Year": "year",
    "Genre": "genre",
    "Main Cast": "cast",
    "Sinopsis": "synopsis",
    "Score": "global_score",
    "Content Rating": "content_rating",
    "Tags": "tags",
    "Network": "network",
    "img url": "img_url",
    "Episode": "episode_raw"
})

# Remove unnamed junk columns if present
kdrama = kdrama.loc[:, ~kdrama.columns.str.contains("^Unnamed")]

kdrama.head()
```

```python
# Convert types
kdrama["year"] = pd.to_numeric(kdrama["year"], errors="coerce")
kdrama["global_score"] = pd.to_numeric(kdrama["global_score"], errors="coerce")
```

```python
# Convert "10 episodes" → 10
kdrama["episodes"] = (
    kdrama["episode_raw"]
    .astype(str)
    .str.extract(r"(\d+)")
    .astype("Int64")
)

kdrama[["title", "episode_raw", "episodes"]].head()
```

```python
def fix_encoding(text):
    if not isinstance(text, str):
        return text
    return (
        text.replace("‚Äú", '"')
            .replace("‚Äù", '"')
            .replace("‚Äô", "'")
            .replace("\\s", "'")
    )

for col in ["title", "synopsis", "genre", "tags", "cast"]:
    kdrama[col] = kdrama[col].apply(fix_encoding)
```

```python
import re

def normalize_title(title):
    if not isinstance(title, str):
        return ""
    
    t = title.lower().strip()
    
    # remove leading a/an/the
    t = re.sub(r'^(a|an|the)\s+', '', t)
    
    # remove punctuation
    t = re.sub(r'[^\w\s]', '', t)

    # collapse multiple spaces
    t = re.sub(r'\s+', ' ', t).strip()
    
    return t
```

```python
kdrama["title_clean"] = kdrama["title"].apply(normalize_title)
my_ratings["title_clean"] = my_ratings["title"].apply(normalize_title)

my_ratings.head()
```

```python
from rapidfuzz import process

# List of all cleaned Kaggle titles
kaggle_titles = kdrama["title_clean"].tolist()

def fuzzy_match(title, choices, threshold=80):
    """Return the best fuzzy match if score >= threshold."""
    match, score, idx = process.extractOne(title, choices)
    if score >= threshold:
        return match
    return None

# Apply fuzzy matching to your titles
my_ratings["title_clean_matched"] = my_ratings["title_clean"].apply(
    lambda x: fuzzy_match(x, kaggle_titles)
)

my_ratings[["title", "title_clean", "title_clean_matched"]].head(20)
```

```python
merged = my_ratings.merge(
    kdrama,
    left_on="title_clean_matched",
    right_on="title_clean",
    how="left",
    suffixes=("_me", "_kaggle")
)
```

```python
total = len(merged)
matched = merged["global_score"].notna().sum()
unmatched = total - matched

print(f"Total shows: {total}")
print(f"Matched with Kaggle: {matched}")
print(f"Not found in Kaggle: {unmatched}")
```

```python
# Work only with matched shows that have genre info
genre_df = matched_df[["title_me", "rating", "global_score", "genre"]].copy()

# Split the comma-separated genre string into a list
genre_df["genre_list"] = (
    genre_df["genre"]
    .fillna("")
    .str.split(",")
    .apply(lambda lst: [g.strip() for g in lst if g.strip() != ""])
)

# Explode so each row = one (show, genre) pair
genre_exploded = genre_df.explode("genre_list")

genre_exploded.head(10)
```

```python
genre_stats = (
    genre_exploded
    .groupby("genre_list")
    .agg(
        my_avg_rating=("rating", "mean"),
        global_avg_score=("global_score", "mean"),
        count=("title_me", "count")
    )
    .sort_values("my_avg_rating", ascending=False)
)

genre_stats.head(20)
```

```python
top_genres = genre_stats[genre_stats["count"] >= 3].copy()  # only genres you've seen at least 3 times

top_genres_sorted = top_genres.sort_values("my_avg_rating", ascending=False)

top_genres_sorted[["my_avg_rating", "global_avg_score"]].head(10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
top_genres_sorted["my_avg_rating"].head(10).plot(kind="bar")
plt.title("My Average Rating by Genre (min 3 shows)")
plt.ylabel("My Avg Rating")
plt.xlabel("Genre")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
# Work only with matched shows that have cast info
actor_df = matched_df[["title_me", "rating", "global_score", "cast"]].copy()

# Split comma-separated cast into list
actor_df["actor_list"] = (
    actor_df["cast"]
    .fillna("")
    .str.split(",")
    .apply(lambda lst: [a.strip() for a in lst if a.strip() != ""])
)

# Explode so each row = one (show, actor) pair
actor_exploded = actor_df.explode("actor_list")

actor_exploded.head(10)
```

```python
actor_stats = (
    actor_exploded
    .groupby("actor_list")
    .agg(
        my_avg_rating=("rating", "mean"),
        global_avg_score=("global_score", "mean"),
        count=("title_me", "count")
    )
    .sort_values(["count", "my_avg_rating"], ascending=[False, False])
)

actor_stats.head(20)
```

```python
favorite_genre_list = fav_genres_strong.index.tolist()
favorite_genre_list
```

```python
# Favorite actors = appeared in at least 2 shows and you rate them highly
fav_actors = (
    actor_stats[actor_stats["count"] >= 2]  # seen enough
    .sort_values("my_avg_rating", ascending=False)
)

# If you want the really top ones
fav_actors_strong = fav_actors[fav_actors["my_avg_rating"] >= 9.0]

favorite_actor_list = fav_actors_strong.index.tolist()
favorite_actor_list[:15]
```

```python
# All titles (cleaned) that you've already rated and matched to Kaggle
watched_titles_clean = (
    my_ratings["title_clean_matched"]
    .dropna()
    .unique()
)

len(watched_titles_clean)
```

```python
candidates = kdrama[~kdrama["title_clean"].isin(watched_titles_clean)].copy()
print("Candidate pool size:", candidates.shape[0])

candidates[["title", "year", "global_score", "genre", "cast"]].head(10)
```

```python
# Ensure genre_list and actor_list columns exist
if "genre_list" not in kdrama.columns:
    kdrama["genre_list"] = (
        kdrama["genre"]
        .fillna("")
        .str.split(",")
        .apply(lambda lst: [g.strip() for g in lst if g.strip() != ""])
    )

if "actor_list" not in kdrama.columns:
    kdrama["actor_list"] = (
        kdrama["cast"]
        .fillna("")
        .str.split(",")
        .apply(lambda lst: [a.strip() for a in lst if a.strip() != ""])
    )

# Make sure candidates has these columns too
candidates["genre_list"] = kdrama.loc[candidates.index, "genre_list"]
candidates["actor_list"] = kdrama.loc[candidates.index, "actor_list"]
```

```python
favorite_genre_set = set(favorite_genre_list)
favorite_actor_set = set(favorite_actor_list)

def count_overlap(values, fav_set):
    return sum(1 for v in values if v in fav_set)

# Calculate overlaps
candidates["genre_overlap"] = candidates["genre_list"].apply(
    lambda lst: count_overlap(lst, favorite_genre_set)
)
candidates["actor_overlap"] = candidates["actor_list"].apply(
    lambda lst: count_overlap(lst, favorite_actor_set)
)

# Normalize global_score to 0–1 (since it's roughly 7–9.5)
candidates["global_score_norm"] = candidates["global_score"] / 10.0

# Simple weighted score:
# tweak weights if you like
candidates["reco_score"] = (
    candidates["global_score_norm"] * 0.5 +
    candidates["genre_overlap"] * 0.3 +
    candidates["actor_overlap"] * 0.2
)
```

```python
top_recos = (
    candidates
    .sort_values("reco_score", ascending=False)
    .head(10)
    [["title", "year", "global_score", "genre", "cast", "genre_overlap", "actor_overlap", "reco_score"]]
)

top_recos
```
